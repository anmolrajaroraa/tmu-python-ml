```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
res = urlopen('https://earthquake.usgs.gov/earthquakes/browse/stats.php')
html = bs(res)
table = html.find('table')
headings = table.find_all('th')
headers = []
for heading in headings:
    headers.append(heading.text)
del headers[0]
rows = headers[-5:]
columns = headers[:-5]
tdList = table.find_all('td')
for i in range(len(tdList)):
    tdList[i] = tdList[i].text
import numpy as np
tdArray = np.array(tdList)
import pandas as pd
tdArray = tdArray.reshape(5,19)
pd.DataFrame(tdArray, index=rows, columns=columns)
```
